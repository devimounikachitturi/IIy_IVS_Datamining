```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\devim\OneDrive\Desktop\Student data.csv")

original_rows = len(df)

# Remove duplicates
df = df.drop_duplicates()
duplicates_removed = original_rows - len(df)

# Standardize column names
df.columns = [c.strip().replace(" ", "_") for c in df.columns]

# Count missing values before cleaning
missing_before = df.isnull().sum().sum()

# Handle missing values
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = df[col].fillna(df[col].median())
    else:
        if df[col].isna().any():
            df[col] = df[col].fillna(df[col].mode()[0])

missing_after = df.isnull().sum().sum()

# Trim text
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()

# Handle outliers (IQR method)
outlier_count = 0
for col in df.select_dtypes(include=['int64', 'float64']):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    before = df[col].copy()
    df[col] = np.where(df[col] < lower, lower,
                np.where(df[col] > upper, upper, df[col]))
    outlier_count += (before != df[col]).sum()

# Save cleaned file
df.to_excel(r"C:\Users\devim\OneDrive\Desktop\Student_data_cleaned.xlsx", index=False)

# Summary
summary = {
    "Original rows": original_rows,
    "Final rows": len(df),
    "Duplicates removed": duplicates_removed,
    "Missing values before": missing_before,
    "Missing values after": missing_after,
    "Outliers capped": int(outlier_count)
}

print("Cleaned dataset shape:", df.shape)
print("Summary of changes:", summary)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load integrated dataset
df = pd.read_excel(r"C:\Users\devim\OneDrive\Desktop\integrated_student_dataset.xlsx")

# Convert date columns to datetime
for col in df.columns:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Encode categorical columns (only low-cardinality to avoid huge size)
cat_cols = [c for c in df.select_dtypes(include='object').columns if df[c].nunique() < 50]
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Normalize numeric columns
num_cols = df_encoded.select_dtypes(include=['int64','float64']).columns
scaler = MinMaxScaler()
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

# Save transformed dataset
df_encoded.to_excel(r"C:\Users\devim\OneDrive\Desktop\transformed_student_dataset.xlsx", index=False)

# 🔹 PRINT OUTPUT
print("\nPreview of Transformed Data:")
print(df_encoded.head())

print("\nTransformed Dataset Shape:", df_encoded.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# Load Datasets
# =========================
attendance = pd.read_csv(r"C:\Users\devim\OneDrive\Desktop\Student attendance.csv")
student = pd.read_csv(r"C:\Users\devim\OneDrive\Desktop\Student data.csv")
homework = pd.read_csv(r"C:\Users\devim\OneDrive\Desktop\student Homework.csv")

# =========================
# Clean & Prepare
# =========================
student.rename(columns={"student_id": "Student_ID"}, inplace=True)
attendance["Attendance_Status"] = attendance["Attendance_Status"].str.strip().str.lower()
homework["Status"] = homework["Status"].str.strip()

# =========================
# Merge All 3 Datasets
# =========================
merged = pd.merge(student, attendance, on="Student_ID", how="left")
merged = pd.merge(merged, homework, on=["Student_ID", "Subject"], how="left")

# =========================
# GRAPH 1 – Subject-wise Present Count
# =========================
present_count = merged[merged["Attendance_Status"]=="present"].groupby("Subject")["Student_ID"].count()

plt.figure()
present_count.plot(kind="bar")
plt.title("Subject-wise Present Students")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# =========================
# GRAPH 2 – Average Marks vs Attendance
# =========================
avg_marks = merged.groupby("Attendance_Status")[["math_score","reading_score","writing_score"]].mean()

plt.figure()
avg_marks.plot(kind="bar")
plt.title("Average Marks vs Attendance")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()

# =========================
# GRAPH 3 – Homework Status Count
# =========================
hw_count = merged["Status"].value_counts()

plt.figure()
hw_count.plot(kind="bar")
plt.title("Homework Status Count")
plt.xlabel("Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# =========================
# GRAPH 4 – Grade vs Avg Math Score
# =========================
grade_avg = merged.groupby("grade_level")["math_score"].mean()

plt.figure()
grade_avg.plot(kind="line")
plt.title("Grade Level vs Average Math Score")
plt.xlabel("Grade Level")
plt.ylabel("Math Score")
plt.tight_layout()
plt.show()
```
